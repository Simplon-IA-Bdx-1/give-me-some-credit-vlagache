```python
import matplotlib.pyplot as plt


auc_av = [0.8327,0.8459,0.8505,0.8533,0.8547,0.8560,0.8560,0.8564,0.8570,0.8572]
ecart_type=[0.006654770434,0.006518451613,0.007065795746,0.006852610764,0.006539064881,0.006655005062,0.006353457102,0.006722501519,0.006643695042,0.006405602979]


auc_less_std = []
auc_plus_std = []
split_data = []

for i in range(0,10):
    auc_less_std.append(auc_av[i] - ecart_type[i])
    auc_plus_std.append(auc_av[i] + ecart_type[i])
    
for i in range(10,101,10):
    split_data.append(i)


plt.plot(split_data,auc_av, ':go' , color='blue')
plt.fill_between(split_data, auc_less_std, auc_plus_std, color='blue', alpha=0.2)
plt.show()
```
