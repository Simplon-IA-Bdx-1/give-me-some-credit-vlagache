```python
import pandas as pd 
from sklearn import metrics
import matplotlib.pyplot as plt


plt.rcParams['figure.figsize'] = [9.5 , 6]
# <Figure size 960x640 with 0 Axes>


df_deep = pd.read_csv('files_csv/GiveMeCredit_Review_Deepnet.csv', index_col=0)
df_ens = pd.read_csv('files_csv/GiveMeCredit_Review_Ensemble.csv', index_col=0)


actual_deep = df_deep['SeriousDlqin2yrs']
scores_deep = df_deep['1 probability']
fpr_deep, tpr_deep, thresholds_deep = metrics.roc_curve(actual_deep, scores_deep, pos_label=1)
auc_deep = metrics.roc_auc_score(actual_deep,scores_deep)

actual_ens = df_ens['SeriousDlqin2yrs']
scores_ens = df_ens['1 probability']
fpr_ens, tpr_ens, thresholds_ens = metrics.roc_curve(actual_ens, scores_ens, pos_label=1)
auc_ens = metrics.roc_auc_score(actual_ens,scores_ens)

plt.figure(figsize=(12,8), dpi=80)
plt.figure()
lw = 2
# plt.plot(fpr,tpr, color='darkorange', lw = lw, label='ROC curve ( area = %0.2f)' %auc)
plt.plot(fpr_deep, tpr_deep, color='darkorange',
         lw=lw, label='ROC curve Deepnet (area = %0.2f)' % auc_deep)
plt.plot(fpr_ens, tpr_ens, color='blue',
         lw=lw, label='ROC curve Ensemble (area = %0.2f)' % auc_ens)
# plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve -  Give Me Credit')
plt.legend(loc="lower right")
plt.show()
```
